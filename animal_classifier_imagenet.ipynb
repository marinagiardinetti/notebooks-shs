```python
# Pas de ré-entraînement. Modèle : imagenet

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions

from google.colab import drive
drive.mount('/content/drive')

# Load MobileNetV2 model trained on ImageNet dataset
model = MobileNetV2(weights='imagenet')

def predict_animal(image_path):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)

    preds = model.predict(img_array)
    predictions = decode_predictions(preds, top=1)[0]
    _, animal_class, confidence = predictions[0]

    return animal_class, confidence

# Example usage
image_path = '/content/drive/My Drive/animal_classifier/elephant.jpg'
animal, confidence = predict_animal(image_path)
print(f"The model predicts that the image contains a {animal} with confidence {confidence}")
```
