```
# @title Chargement du corpus

#@markdown Dossier de travail:
working_dir = '/content/drive/My Drive/NER_BNF/' #@param {type:"string"}
os.chdir(working_dir)

#@markdown Dossier contenant le corpus de textes
corpus = '/content/drive/My Drive/NER_BNF/corpus2/' #@param {type:"string"}

# Afficher le nombre de textes

textes = os.listdir(corpus)
print('Nombre de textes chargés dans le corpus: ', len(textes))


  # Notions d'histoire générale et histoire de France depuis l'antiquité jusqu'à nos jours / par MM. A. Aulard,... A. Debidour
  # bpt6k34118531 n'a pas été océrisé.
```

```
# @title Recherche et extraction des entités nommées

model.max_length = 150000000

nom_glob = ''.join(corpus+"*.txt")
for element in glob.glob(nom_glob):
  with open (element, "r") as myfile:
    data=myfile.read()
    doc = model(data)

    liste_lieux = []
    for ent in doc.ents:
          if ent.label_ == 'LOC':
            liste_lieux.append(ent.text)
            df_loc = pd.DataFrame(liste_lieux)
            df_loc.to_excel(''.join([working_dir, 'spacy_lieux.xlsx']))

    liste_pers=[]
    for ent in doc.ents:
          if ent.label_ == 'PER':
            liste_pers.append(ent.text)
            df_pers = pd.DataFrame(liste_lieux)
            df_pers.to_excel(''.join([working_dir, 'spacy_pers.xlsx']))
```
